# Sample meta fields

```python
from elabjournal import elabjournal
eLAB = elabjournal.api()
```

```python
#always use a specific group for this example
eLAB.set_group(6727)
eLAB.group()
```

## Get sample
Test sample

```python
testSample = eLAB.sample(7627586)
testSample.visualize()
```

## Get meta fields for sample

```python
testSampleMetas = testSample.metas()
testSampleMetas
```

```python
testSampleMeta = testSampleMetas.first()
testSampleMeta.visualize()
```

```python
testSampleMeta.data()
```

## Get sample type

```python
sampleType = testSample.sample_type()
sampleType.visualize()
```

## Get meta fields for sample type

```python
sampleTypeMetas = sampleType.metas()
sampleTypeMetas.all()
```

```python
sampleTypeMeta = sampleTypeMetas.first()
display(sampleTypeMeta.data())
sampleTypeMeta.visualize()
```

## Update meta fields

```python
metasList = testSampleMetas.all()
metasList
```

```python
metaText = testSample.meta(metasList[metasList["sampleDataType"]=="TEXT"].index.values[0])
display(metaText.data())
oldKey = metaText.key()
oldValue = metaText.value()
metaText.update(key="testkey 123", value="testvalue 456")
display(metaText.data())
metaText.update(key=oldKey, value=oldValue)
display(metaText.data())
```

```python
metaRadio = testSample.meta(metasList[metasList["sampleDataType"]=="RADIO"].index.values[0])
metaRadioType = sampleType.meta(metaRadio.data()["sampleTypeMetaID"])
display(metaRadio.data())
oldKey = metaRadio.key()
oldValue = metaRadio.value()
oldSampleTypeMetaID = metaRadio.data()["sampleTypeMetaID"]
options = list(metaRadioType.data()["optionValues"])
options.remove(oldValue)
newValue = options[0]
metaRadio.update(key="testkey 123", value=newValue, sampleTypeMetaID=0)
display(metaRadio.data())
metaRadio.update(key=oldKey, value=oldValue, sampleTypeMetaID=oldSampleTypeMetaID)
display(metaRadio.data())
```
