# Sample meta fields

```python
from elabjournal import elabjournal
eLAB = elabjournal.api()
```

```python
#always use a specific group for this example
eLAB.set_group(6727)
eLAB.group()
```

## Get sample
Test sample

```python
testSample = eLAB.sample(7627586)
testSample.visualize()
```

## Get meta fields for sample

```python
testSampleMetas = testSample.metas()
testSampleMetas
```

```python
testSampleMeta = testSampleMetas.first()
testSampleMeta.visualize()
```

```python
testSampleMeta.data()
```

```python
sampleType = testSample.sample_type()
sampleType.visualize()
```

```python
sampleTypeMeta = sampleTypeMetas.first()
sampleTypeMeta.visualize()
```
